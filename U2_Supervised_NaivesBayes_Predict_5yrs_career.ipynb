#**Step 3: Feature engineering**

**Question:** Which columns would you select and avoid selecting as predictors for the target variable, `target_5yrs`?

- Avoid the `unamed` and `name` column since they are not helpful
- The number of games a player has play, `gp`,  can be combined with `pts` to get the **total number of points** their earned across all games
- Also combine `gp` and `pts` with `min` to extract another feature, which is the **player's efficiency**
- Drop some similar columns: keep **percentage of field goals** `fg` and drop `fgm` and `fga`, keep **number of rebounds overal**l `reb` and drop  `oreb` and `dreb`
- Keep **number of assists** `ast`, **steals** `slt`, **blocks** `blk`, and **turnovers** `tov`.  

Therefore, we will use the following columns
`gp`, `min`, `pts`, `fg`, `3p`, `ft`, `reb`, `ast`, `stl`, `blk`, `tov`

```python
# Select the columns to proceed with and extract two features: total points and efficiency

selected_data = data[['gp', 'min', 'pts', 'fg', '3p', 'ft', 'reb', 'ast', 'stl', 'blk', 'tov', 'target_5yrs']]
selected_data['total_points'] = selected_data['gp'] * selected_data['pts']
selected_data['efficiency'] = selected_data['pts'] / selected_data['min']
selected_data.head()
```

```python
# Drop columns `gp`, `pts`, and  `min`
selected_data.drop(['gp', 'pts', 'min'], axis=1, inplace=True)
selected_data.head()
```

```python
# Export the extracted data to a new .csv file
selected_data.to_csv('/content/gdrive/MyDrive/Colab Notebooks/ML Coursera/nba-players-extracted.csv', index=False)
```
